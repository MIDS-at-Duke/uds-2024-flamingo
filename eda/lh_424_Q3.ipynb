### Monthly plot for each sector from 1973/01 - 2023/10

```python
# Exclude annual data, which is "YYYY13" in the YYYYMM column

df1["YYYYMM_str"] = df1["YYYYMM"].astype(str)
df2["YYYYMM_str"] = df2["YYYYMM"].astype(str)

df1_monthly = df1[~df1["YYYYMM_str"].str.endswith("13")]
df2_monthly = df2[~df2["YYYYMM_str"].str.endswith("13")]

df1_monthly = df1_monthly.drop("YYYYMM_str", axis=1)
df2_monthly = df2_monthly.drop("YYYYMM_str", axis=1)

df1_monthly["Date"] = pd.to_datetime(df1_monthly["YYYYMM"], format="%Y%m")
df2_monthly["Date"] = pd.to_datetime(df2_monthly["YYYYMM"], format="%Y%m")

print(df1_monthly.shape)
df1_monthly.head()
```

```python
# Monthly data for each sector

residential = df1_monthly[df1_monthly["Column_Order"] == 5]
commercial = df1_monthly[df1_monthly["Column_Order"] == 10]
industrial = df1_monthly[df1_monthly["Column_Order"] == 15]
transportation = df2_monthly[df2_monthly["Column_Order"] == 5]

residential.head()
```

```python
plt.figure(figsize=(14, 8))

# Residential
plt.subplot(2, 2, 1)
plt.plot(
    residential["Date"],
    residential["Value"].astype(float),
    label="Residential",
    color="blue",
)
plt.title("Residential Sector Energy Consumption")
plt.xlabel("Time")
plt.ylabel("Energy Consumption (Trillion Btu)")
plt.xticks(rotation=45)
plt.tight_layout()

# Commercial
plt.subplot(2, 2, 2)
plt.plot(
    commercial["Date"],
    commercial["Value"].astype(float),
    label="Commercial",
    color="green",
)
plt.title("Commercial Sector Energy Consumption")
plt.xlabel("Time")
plt.ylabel("Energy Consumption (Trillion Btu)")
plt.xticks(rotation=45)
plt.tight_layout()

# Industrial
plt.subplot(2, 2, 3)
plt.plot(
    industrial["Date"],
    industrial["Value"].astype(float),
    label="Industrial",
    color="red",
)
plt.title("Industrial Sector Energy Consumption")
plt.xlabel("Time")
plt.ylabel("Energy Consumption (Trillion Btu)")
plt.xticks(rotation=45)
plt.tight_layout()

# Transportation
plt.subplot(2, 2, 4)
plt.plot(
    transportation["Date"],
    transportation["Value"].astype(float),
    label="Transportation",
    color="purple",
)
plt.title("Transportation Sector Energy Consumption")
plt.xlabel("Time")
plt.ylabel("Energy Consumption (Trillion Btu)")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

### Residential Sector monthly plot for 2021 - 2023

```python
residential["Year"] = residential["Date"].dt.year
years_interest = [2021, 2022, 2023]
residential_recent_years = residential[residential["Year"].isin(years_interest)]

plt.figure(figsize=(10, 6))
plt.plot(
    residential_recent_years["Date"],
    residential_recent_years["Value"].astype(float),
    marker="o",
    linestyle="-",
    color="green",
)
plt.title("Residential Sector Monthly Energy Consumption (2021-2023)")
plt.xlabel("Month")
plt.ylabel("Total Energy Consumption (Trillion Btu)")
plt.xticks(residential_recent_years["Date"])
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Percentage consumption of sectors in Oct 2023

```python
residential_oct_2023 = residential[residential["YYYYMM"] == 202310]["Value"].sum()
commercial_oct_2023 = commercial[commercial["YYYYMM"] == 202310]["Value"].sum()
industrial_oct_2023 = industrial[industrial["YYYYMM"] == 202310]["Value"].sum()
transportation_oct_2023 = transportation[transportation["YYYYMM"] == 202310][
    "Value"
].sum()

values = [
    residential_oct_2023,
    commercial_oct_2023,
    industrial_oct_2023,
    transportation_oct_2023,
]
labels = ["Residential", "Commercial", "Industrial", "Transportation"]

plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct="%1.1f%%", startangle=140)
plt.title("Energy Consumption by Sector in October 2023")
plt.show()
```

### Percentage of sectors across months

```python
df_combined = pd.concat(
    [
        residential.assign(Sector="Residential"),
        commercial.assign(Sector="Commercial"),
        industrial.assign(Sector="Industrial"),
        transportation.assign(Sector="Transportation"),
    ]
)

monthly_totals_by_sector = (
    df_combined.groupby(["Date", "Sector"])["Value"].sum().reset_index()
)
monthly_totals_by_sector.head()
```

```python
total_monthly_energy = (
    monthly_totals_by_sector.groupby("Date")["Value"]
    .sum()
    .reset_index(name="TotalEnergy")
)
```

```python
monthly_totals_by_sector = monthly_totals_by_sector.merge(
    total_monthly_energy, on="Date"
)
monthly_totals_by_sector["Percentage"] = (
    monthly_totals_by_sector["Value"] / monthly_totals_by_sector["TotalEnergy"]
) * 100

monthly_totals_by_sector.head()
```

```python
monthly_percentages_pivot = monthly_totals_by_sector.pivot(
    index="Date", columns="Sector", values="Percentage"
)
monthly_percentages_pivot
```

```python
plt.figure(figsize=(14, 8))

for sector in monthly_percentages_pivot.columns:
    plt.plot(
        monthly_percentages_pivot.index, monthly_percentages_pivot[sector], label=sector
    )

plt.title("Sector Percentage of Total Energy Consumption Across Months")
plt.xlabel("Date")
plt.ylabel("Percentage of Total Energy Consumption (%)")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Percentage of sectors across months (2021 - 2023)

```python
df_combined_years_2021 = df_combined[
    df_combined["Date"].dt.year.isin([2020, 2021, 2022])
]

monthly_2021 = (
    df_combined_years_2021.groupby(["Date", "Sector"])["Value"].sum().reset_index()
)

total_monthly_2021 = (
    monthly_2021.groupby("Date")["Value"].sum().reset_index(name="TotalEnergy")
)

monthly_2021 = monthly_2021.merge(total_monthly_2021, on="Date")

monthly_2021["Percentage"] = (monthly_2021["Value"] / monthly_2021["TotalEnergy"]) * 100

monthly_pivot_2021 = monthly_2021.pivot(
    index="Date", columns="Sector", values="Percentage"
)

plt.figure(figsize=(14, 8))
for sector in monthly_pivot_2021.columns:
    plt.plot(
        monthly_pivot_2021.index, monthly_pivot_2021[sector], label=sector, marker="o"
    )

plt.title("Sector Percentage of Total Energy Consumption Across Months (2021-2023)")
plt.xlabel("Date")
plt.ylabel("Percentage of Total Energy Consumption (%)")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
