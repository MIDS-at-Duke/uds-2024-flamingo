```python
import pandas as pd

pd.set_option("mode.copy_on_write", True)

df = pd.read_excel("../src/Prod_dataset.xlsx")
```

```python
df_melt = df.drop(columns=["Data_Status"]).melt(
    id_vars=["StateCode", "MSN"], var_name="Year", value_name="Value"
)
data = df_melt.pivot(index=["StateCode", "Year"], columns="MSN", values="Value")
data = data.reset_index()
data
```

```python
data["Fuels_ratio"] = (data["CLPRB"] + data["NGMPB"] + data["PAPRB"]) / data["TEPRB"]
```

```python
import matplotlib.pyplot as plt


def plot_fossil_fuel_ratio_trend(data, year, treated, control):
    df_treated = data[(data["StateCode"] == treated) & (data["Year"] >= year)]
    df_control = data[(data["StateCode"] == control) & (data["Year"] >= year)]
    plt.plot(
        df_treated["Year"],
        df_treated["Fuels_ratio"],
        c="b",
        label=f"treated ({treated})",
    )
    plt.plot(
        df_control["Year"],
        df_control["Fuels_ratio"],
        c="r",
        label=f"control ({control})",
    )
    plt.xlabel("Year")
    plt.ylabel("Fossil Fuels Ratio")
    plt.title("State Fossil Fuels Ratio Trend")
    plt.legend()
    plt.show()
```

```python
plot_fossil_fuel_ratio_trend(data, 2000, "MD", "VA")
```

```python
plot_fossil_fuel_ratio_trend(data, 2000, "DE", "PA")
```
