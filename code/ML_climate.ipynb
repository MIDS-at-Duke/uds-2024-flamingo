```python
import pandas as pd

pd.set_option("mode.copy_on_write", True)

df_cddcst = pd.read_fwf(
    "../src/climdiv-cddcst-v1.0.0-20240404.txt",
    widths=[3, 1, 2, 4, 7, 7, 7, 7, 7, 7, 7, 7, 7, 7, 7, 7],
    header=None,
)
df_hddcst = pd.read_fwf(
    "../src/climdiv-hddcst-v1.0.0-20240404.txt",
    widths=[3, 1, 2, 4, 7, 7, 7, 7, 7, 7, 7, 7, 7, 7, 7, 7],
    header=None,
)
```

```python
df_cddcst["cdd"] = df_cddcst.iloc[:, 4:].mean(axis=1)
df_cddcst.describe()
```

```python
df_cdd = df_cddcst.iloc[:, [0, 3, -1]]
df_cdd.columns = ["state_code", "Year", "cdd"]
df_cdd.describe()
```

```python
df_hddcst["hdd"] = df_hddcst.iloc[:, 4:].mean(axis=1)
df_hdd = df_hddcst.iloc[:, [0, 3, -1]]
df_hdd.columns = ["state_code", "Year", "hdd"]
df_hdd.describe()
```

```python
df_cddhdd = pd.merge(df_cdd, df_hdd, on=["state_code", "Year"])
df_cddhdd.describe()
```

```python
import numpy as np

df_cddhdd = df_cddhdd[df_cddhdd["state_code"] < 100]
df_cddhdd = df_cddhdd[(df_cddhdd["Year"] <= 2020) & (df_cddhdd["Year"] >= 2000)]
df_cddhdd.replace(to_replace=-9999, value=np.nan, inplace=True)
```
