```python
import pandas as pd

pd.set_option("mode.copy_on_write", True)

df = pd.read_csv("../src/Average_retail_price_of_electricity_2001-2023.csv", skiprows=4)
```

```python
df.head(10)
```

```python
df.info()
```

```python
df = df.dropna().reset_index(drop=True)
df.head(10)
```

```python
df["state"] = df["source key"].str.replace("ELEC.PRICE.", "").str.replace("-ALL.A", "")
df.head(10)
```

```python
df.drop(columns=["source key", "description", "units", "source key"], inplace=True)
```

```python
df["state"].unique()
```

```python
df[df["state"].str.contains("AT")]
```

```python
AT_row = df[df["state"].str.contains("AT")].mean(axis=0)
AT_row["state"] = "AT"
df = df.append(AT_row, ignore_index=True)
```
